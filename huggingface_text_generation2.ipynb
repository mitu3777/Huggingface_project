```python
import pandas as pd
poems = pd.read_csv('/robert_frost_collection.csv')
poems.head(5)
```

```python
content = poems['Content'].dropna().tolist()
```

```python
lines = []
for poem in content:
    for line in poem.split("\n"):
        lines.append(line.rstrip())
```

```python
lines = [line for line in lines if len(line)> 0]
lines[:5]
```

```python
!pip install transformers
from transformers import pipeline

gen = pipeline("text-generation")
```

```python
lines[0]
```

```python
gen(lines[0], max_length=20, num_return_sequences=3)
```

```python
import textwrap
def wrap(x):
    return textwrap.fill(x, replace_whitespace=False, fix_sentence_endings=True)
```

```python
out = gen(lines[0], max_length=30)
print(wrap(out[0]['generated_text']))
```

```python
prompt = "Transformers have a wide variety of application in NLP"
OUT = gen(prompt, max_length=100)
print(wrap(OUT[0]['generated_text']))
```
